# MetaGrid

Metagrid is a wrapper class around a pygeogrids Basic/CellGrid that that allows defining, activating and combining multiple subsets for a single grid. Only one subset can be active for a grid at a time. Subsets in a Metagrid can be combined (intersection, union, diff, ...) to create new subsets. MetaGrids can be loaded from and saved to netcdf files.

## Basic Example
The first examples shows the basic functionality of MetaGrid for a simple 9x9 DEG global grid.

```python
from pygeogrids import MetaGrid
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

lon, lat = np.meshgrid(np.arange(-180.,180.,9.), np.arange(-90.,90.,9.))
metagrid = MetaGrid(lon.flatten(), np.flipud(lat).flatten())
metagrid.plot(False, markersize=5.)
```

 Then we add add two subsets from bounding boxes and a third one from their intersection.

```python
metagrid.subset_from_bbox(-60,60,-50,50, name='center_box')
metagrid.subset_from_bbox(-20,20,-160,160, name='lon_box')
metagrid.combine_subsets(['lon_box', 'center_box'], method='intersect', new_name='inter')
```

The three subsets look like this

```python
fig, axs = plt.subplots(1,3, figsize=(15,8), subplot_kw={'projection': ccrs.Orthographic(0,0)})
metagrid.activate_subset('center_box')
metagrid.plot(only_subset=False, markersize=15., title='center_box', color='blue', ax=axs[0], set_auto_extent=False)
metagrid.activate_subset('lon_box')
metagrid.plot(only_subset=False, markersize=15., title='lon_box', color='red', ax=axs[1], set_auto_extent=False)
metagrid.activate_subset('inter')
metagrid.plot(only_subset=False, markersize=15., title='inter', color='orange', ax=axs[2], set_auto_extent=False)
```

Now we can combine the two subsets (intersection) to create a third subset.

## Real data example
For this example we create a local quarter degree grid and load an according pre-defined SubsetCollection from file.

```python
try:
    import matplotlib.pyplot as plt
    import cartopy
    makemaps= True
except ImportError:
    makemaps = False
    print('Parts of this tutorial need matplotlib and cartopy installed.')
    print("Run 'conda install cartopy'")
import numpy as np
import os
from pprint import pprint
from pygeogrids import src_path
from pygeogrids.subset import SubsetCollection
from pygeogrids.grids import genreg_grid

base_grid = genreg_grid(grd_spc_lat=0.25, grd_spc_lon=0.25, origin='bottom')
subsets = SubsetCollection.from_file(os.path.join(src_path, '..', 'docs', 'examples', 'metagrid', 'europe.nc'))
```

The subsets in the loaded collection contain a flags for land point (binary, land/not land) over Europe as well as non-binary subsets that indicate the country and landcover type of a grid point in Europe.

Collection contains the follwoing subsets:

```python
subsets.names
```

Subset 'land' is a binary subset (mask) of land points (i.e. all values are 1) in the selected bounding box:

```python
subsets['land'].as_dict()
```

Subset 'country' is a non-binary subset (classes) of points within each country in the bounding box around Europe.

```python
print(subsets['country'].as_dict())
```

We then create a MetaGrid object from the existing grid, and also use the loaded SubsetCollection.

```python
from pygeogrids.grids import MetaGrid
metagrid = MetaGrid.from_grid(base_grid, subsets=subsets)
```

We can now activate, combine, merge, visualise and use different subsets in the grid. We start by activating the land mask. And visualise the land points on a map. In this plot, ALL (global) points are show in gray, and the subset points are plotted in green.

```python
metagrid.activate_subset('land')
if makemaps:
    fig, axs = plt.subplots(1,2, figsize=(12,8), subplot_kw={'projection': ccrs.Mercator()})
    metagrid.plot(only_subset=False, title='Europe Land Mask Subset', markersize=1.0, ax=axs[0])
    metagrid.plot(only_subset=True, title='Europe Land Mask Subset', markersize=1.0, ax=axs[0])
    plt.show()
else:
    print('Install matplotlib and cartopy to run this part of the tutorial')
```

The same for the other two (non-binary) subsets, without the non-subset grid points:

```python
metagrid.activate_subset('country')
if makemaps:
    metagrid.plot(only_subset=True, visualise_vals=True, title='Europe Country Name Subset', markersize=1.0)
    plt.show()
else:
    print('Install matplotlib and cartopy to run this part of the tutorial')
```

```python
metagrid.activate_subset('landcover_class')
if makemaps:
    metagrid.plot(only_subset=True, visualise_vals=True, title='Europe CCI Landcover Subset', markersize=1.0)
    plt.show()
else:
    print('Install matplotlib and cartopy to run this part of the tutorial')
```

### Combine, merge, filter existing subsets.
Now we can combine, merge and filter the available subsets. We can also add new ones.

First we create a new subset by filtering the points for 'France' (id=56 as defined in the attributes of the subset) and add it to the metagrid.

```python
# Filter subset for France and add it to the metagrid
idx = subsets['country'].attrs['flag_meanings'].index('France')
val_france = subsets['country'].attrs['flag_values'][idx]
metagrid.activate_subset('country')
metagrid.filter_active_subset(vals=[val_france], name='france')
print(metagrid.subsets.names)
```

We then do the same for the points that are marked as 'grassland', we use the following LC classes: 120, 121, 122, 130.

```python
for idx in [120, 121, 122, 130]:
    idx = subsets['landcover_class'].attrs['flag_values'].aslist().index(idx)
    meaning = subsets['landcover_class'].attrs['flag_meanings'][idx]
    print(idx, meaning)
```

```python
idx = subsets['country'].attrs['flag_meanings'].index('France')
val_france = subsets['country'].attrs['flag_values'][idx]
metagrid.activate_subset('country')
metagrid.filter_active_subset(vals=[val_france], name='france')
print(metagrid.subsets.names)
```
